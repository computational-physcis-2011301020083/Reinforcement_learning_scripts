```python
import random
import gym,os
import numpy as np
from collections import deque
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten, Conv2D, MaxPooling2D
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(100, input_dim=4, activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(2, activation='linear'))
model.compile(loss='mse', optimizer=Adam(lr=0.001))
```

```python
model.predict([[1,2,3,4]])[0][1]
```

```python
def Act(st,ex,mo1):
    if np.random.rand(1)<= ex:
        a=np.random.randint(2)
        #exp=exp*(1-exp_dec)
    else :
        a=np.argmax(mo1.predict(st)[0])
    return a
```

```python
env=gym.make("CartPole-v1",render_mode="human")
mem=deque(maxlen=10000)
exp=1
exp_dec=0.99975
for i in range(5000):
    print(i,exp)
    state=env.reset()
    done=False
    #action=0
    #print(state)
    while not done:
    #for j in range(2):
        env.render()
        action=Act(np.reshape(state[0], [1, 4]),exp,model)
        #exp=exp*(1-exp_dec)
        next_state,reward,done,_,info=env.step(action)
        #print(env.step(action))
        #print(exp,state[0].shape,state[0])
        #print("state",state[0],state[1],next_state)
        #np.reshape(next_state, [1, 4])
        mem.append((np.reshape(state[0], [1, 4]),action,reward,done,np.reshape(next_state, [1, 4])))
        state=(next_state,{})
    if len(mem)<=64:
        pass
    else:
        batch1=random.sample(mem,64)
        for stat,actio,rewar,don,next_stat in batch1:
            #print(111)
            #print(stat,actio,rewar,don,next_stat)
            target=rewar+0.99*max(model.predict([next_stat])[0])
            target1=model.predict([stat])
            target1[0][actio]=target
            model.fit(stat, target1, epochs=1, verbose=0)
    if exp>=0.001:
        exp=exp*exp_dec
#print(mem)
```
